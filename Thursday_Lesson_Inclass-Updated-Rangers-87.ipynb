##### Accepting Parameters

```python
first_name = "Garrett"
last_name = "Waters"

def print_full_name(first,last):
    return f"Hello my name is {last} and my first name is {first}"

print(print_full_name(last_name,first_name))
```

##### Default Parameters

```python
# Default parameters must come after non-default parameters at all times

def june_bday(day, year, month="June"):
    return f"Your birthday is the {day}th day of {month} and you were born in {year}."

print(june_bday(5,2022))
```

##### Making an Argument Optional

```python
# an argument that defaults to nothing
def print_horse_name(first, middle="", last="Ed"):
    return f"Hello {first} {middle} {last}."
print(print_horse_name("Mr."))
print(print_horse_name("Mr.","Seabiscut"))
```

##### Keyword Arguments

# Creating a start, stop, step function

```python
def print_hero(name, power= "flying"):
    return f"{name}'s power is {power}!"

print(print_hero("Bruce"))
```

```python
# Create a function that takes at least 1 postional, 1 optional, and 1 default argument.
def print_tea(name, tea ="Green Tea", place=""):
    if place:
        return f"{name} likes {tea} at {place}"
    else:
        return f"{name} likes {tea}"
print(print_tea("Garrett"))
```

##### Returning Values

```python
def add_nums(num1,num2):
    return num1 + num2
total = add_nums(35,65)

print(total)
```

##### *args / **kwargs (keyword arguments)

```python
#*args stands for arguments( **kwargs for keyword arguments) & takes any number of arguments as parameters
# must be last if other arguments are present

def print_args(num1, *args, **keywords):
    print(num1)
    print(args)
    print(keywords)

print_args(1,2,3,"icecream", names =["Ryan", "Terrell"], subjet = "Python")
```

```python
def test(loop,*args,**kwargs,):
    for i in range(loop):
        print(args)
        print(kwargs)

test(1000,1,2,3,4,5,"Hi", secret_list =["secret", "secret", "extra secret"])
```

### Exercise 3
Create a function that takes in a list of strings and filters out the strings that have less that six characters . 

```python
string_list = ['Sheldon','Pnny','Leonard','Howard','Raj','Amy','Stuart']
# output = ['Sheldon','Leonard','Howard','Stuart']
def new_string():
    return([i for i in string_list if len(i) >= 6])
print(new_string())
```

### Exercise 4
Create a function that accepts a list as a parameter and returns a dictionary containing the list items as it's keys, and the number of times they appear in the list as the values

```python
example_list = ["Harry", 'Hermione','Harry','Ron','Dobby','Draco','Luna','Harry','Hermione','Ron','Ron','Ron']

# output = {
#     "Harry":3,
#     "Hermione":2,
#     "Ron":4,
#     "Dobby":1,
#     "Draco":1,
#     "Luna": 1
# }
def dict_counted():
    new_dict = {i: example_list.count(i) for i in example_list}
    return(new_dict)
print(dict_counted())
```

<p>Scope refers to the ability to access variables, different types of scope include:<br>a) Global<br>b) Function (local)<br>c) Class (local)</p>

```python
# placement of variable declaration matters

number = 3 # Gloal Variable

def myFunc():
    num_3 = 6 # Local Function Variable
    return num_3

print(number)
return_num = myFunc()
print(return_num)
```
